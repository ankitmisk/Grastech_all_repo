# Private

```python
class Audi:

    def __init__(self): # class constructor
        self.__wheel = 4
        self.color = 'black'
        self.engine = 'V8-Engine'
        # private instance variable
        self.__digital_key = True
        self.__air_bag = 9
        print('You car is fully Loaded!!')

    # Methods
    def auto_pilot(self,off = True):
        if off == True:
            print('Car is in Mannual Mode!!')
        else:
            print('Auto Mode ON!!!')

    def engine_start(self):
        print('Engine Start!!')
        import pygame
        pygame.init()

        music = pygame.mixer.Sound(r"C:\Users\salma\Downloads\lambo-start-up-sound-26364.mp3")
        music.play()

        
    def engine_stop(self):
        print('Engine Stop!!')
    # Private method
    def __Hazard_(self):
        print('Emergency Light ON!!')
        
    # Private method  
    def __check_update_and_system_config(self):
        print('Checking system updates and config!!')
```

```python
r10 = Audi()
```

```python
r10._Audi__digital_key
```

```python
r10._Audi__Hazard_()
```

```python
r10.engine_start()
```

# Multiple Inheritance

```python
class Exide:
    def super_charged_battery(self):
        print('Very fast auto charged battery!!')

class Samsung:
    def display_4k(self):
        print('4k quality Display!!')

class Boat:
    def headphone(self):
        print('JBL speaker Headphone!!')

class Sony:
    def camera(self):
        print('64 Mega Pixel Camera, Auto-focus')

class Software_and_hardware_features:
    def strong(self):
        print('Very Strong material')
        
    def multiple_call(self):
        print('Multiple call')
        
    def sim_5g(self):
        print('Supports 5g Sim')
    def ringtone_selection(self):
        print('Can change Ringtones')
```

```python
class Nokia(Samsung,Boat,Exide, Sony, Software_and_hardware_features):
    def Mobile_case(self):
        print('Very Strong Casing!!')
```

```python
nokia_1100 = Nokia()
```

```python
nokia_1100.sim_5g()
```

```python
nokia_1100.ringtone_selection()
```
