```python
def factorial_loop(number):
    result = 1
    for i in range(1,number+1):
        result *= i
    return result
```

```python
factorial_loop(10)
```

```python
def facto_rec(n):
    if n<=1:
        return n
    else:
        return n * facto_rec(n-1)
```

```python
facto_rec(5)
```

```python
# Decorator:- existing function decorate karna
```

```python
def decorate_this(f):
    print('You have been decorated')
    f()
    print('Above Func is original func and it\'s been decorated')
```

```python
def old():
    print('This is just an OLD function')
    print('Let\'s Decorate me')
```

```python
old()
```

```python
decorate_this(old)
```

```python
old()
```

```python
@decorate_this
def old():
    print('This is just an OLD function')
    print('Let\'s Decorate me')
```

```python
@decorate_this
def bye():
    print('Hello Bye bye!!')
```

```python
# map:- numerical mapping
```

```python
data = [3,45,5,5,67,7,3]
data_final = []
for i in data:
    data_final.append(i**2)
print(data_final)
```

```python
data = (3,5,5,5,67,7,3)
```

```python
list(map(lambda n: n**2,data))
```

```python
tuple(map(lambda n: n**2,data))
```

```python
# filter :- filter out the result
```

```python
data = [34,54,5,34,56567,4,5,35,7,6756,45,7,5]
```

```python
even = []
for i in data:
    if i % 2 == 0:
        even.append(i)
print(even)
```
