```python
# Find Common Elements
```

```python
data1 = [4,54,534,5,6,34,554,665,7435]
data2 = [34,15,63,45,534,657,4,5,456]

common_list = []

for i in data1:
    for j in data2:
        if i == j:
            common_list.append(j)
            # print(i,j)

print('Common Elements',common_list)
```
